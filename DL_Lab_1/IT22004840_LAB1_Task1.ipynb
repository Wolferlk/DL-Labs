```python
import matplotlib.pyplot as plt

# Large datasets
exp_data = np.random.exponential(scale=1.0, size=(100000,))
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plot
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.axis([-2.5, 5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create X and Y meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# 3D plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

# Import the pandas library and assign it the alias 'pd'
import pandas as pd

# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# You can now proceed to use the 'df' DataFrame
print(df.head())
```

```python
# Assuming the DataFrame is loaded as 'df'
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Select the specified features
features_df = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# 2. Calculate Pearson correlation
pearson_corr = features_df.corr(method='pearson')

# 3. Visualize Pearson correlation with a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Standard Coefficient")
plt.show()

# 4. Calculate Spearman correlation
spearman_corr = features_df.corr(method='spearman')

# 5. Visualize Spearman correlation with a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Spearman Rank Correlation")
plt.show()
```
